```python
import pandas as pd
import folium
from folium.plugins import HeatMap


df = pd.read_csv('../data/BO_2011_2_tratado.csv')


df = df.head(10000)


mapa = folium.Map(location=[-23.55, -46.63], zoom_start=11)


HeatMap(
    data=df[['LATITUDE','LONGITUDE']],
    radius=8,
    blur=12,
    max_zoom=13
).add_to(mapa)


#mapa.save("heatmap_2011_2_10000.html")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os


os.makedirs('../outputs', exist_ok=True)

contagem_crimes = df['RUBRICA'].value_counts()


contagem_crimes = contagem_crimes[contagem_crimes > 100]

cores = plt.cm.tab20.colors  

plt.figure(figsize=(10,6))
contagem_crimes.plot(kind='barh', color='red')
plt.xlabel('Quantidade de Crimes')
plt.ylabel('Tipo de Crime')
plt.title('Distribuição de Crimes em 2011')
plt.gca().invert_yaxis()  


plt.savefig('../outputs/distribuicaoCrimes2011.png', dpi=300, bbox_inches='tight')

plt.show()
```
